## Divide  Categorical and Continuous Columns

```python
cat=[]
con=[]
for i in A:
    if A[i].dtypes=="object":
        cat.append(i)
    else:
        con.append(i)
```

## Standarization using StandarScaler

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X2=pd.DataFrame(ss.fit_transform(A[con]),columns=con)
X2.head(3)
```

## Standarization usind MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
X2=pd.DataFrame(mm.fit_transform(A[con]),columns=con)
```
